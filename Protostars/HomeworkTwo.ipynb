## Problem 3 Part d

Plot S versus x at T = 1, 1.5, 2, and 4. Give a physical interpretation of the results.

$S = \frac{C}{3 \pi v_1 \Sigma_1} \frac{e^{-x/T}}{x T^{3/2}}$

```python
def surface_density(C, v_1, Sigma_1, x, T):
    return C / (3*np.pi*v_1*Sigma_1) * (np.exp(-x/T) / (x * T**(3/2)))
```

```python
x = np.linspace(0.5, 6, num=100)
T = [1, 1.5, 2, 4]
C = 1
v_1 = 1
Sigma_1 = 1

plt.yscale('log')
#plt.xscale('log')

for time in T:
    S = np.zeros(len(x))
    for i in range(len(x)):
        S[i] = surface_density(C, v_1, Sigma_1, x[i], time)
        
    label = "T = " + str(time)
    plt.plot(x, S, label=label)

plt.xlabel('x [unitless radius]')
plt.ylabel('S [unitless surface density]')
plt.legend()
```

The surface density of the disk is very high near the center, which makes sense as it is where the majority of the mass is centered. Over time, the disk begins to level out so that the inner and outer parts of the disk are closer to the same density. Most of the mass that was in the center of the disk has been accreted, and is going towards equilibrium with the rest of the disk. It is still more dense towards the center of the disk relative to the edges.
