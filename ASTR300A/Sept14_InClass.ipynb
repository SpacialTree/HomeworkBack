## 2.2. Lambda Functions
There are a few basic things that you need to know how to do before we get to our main task for today.  You need to know how to make a function that returns a value, and you need to know how to make and save a plot. 

We can create a `lambda function`, which is a tool in Python to return the answer to an equation quickly and easily. Here's an example of a lambda function for a simple polynomial $y(x)=ax^2+bx+c$:

```python
# This is a lambda function: ('y' is the name of the function.)
# After typing lambda, give the parameters you want to pass into the function.  
# Then, use those variables to calculate the value you want to return.
# NOTE: a power in Python is given by **, not ^ !
y = lambda x,a,b,c: a*x**2 + b*x + c

print(y(0.,3.,-2.,-5.)) # this returns y(0) with a=3, b=-2, and c=-5

# We can also create an array of x values and pass that to our lambda function.
my_x = array([0.,1.,2.,3.,4.])
my_y = y(my_x,3.,-2.,-5.)
print(my_y)
```

## 2.3. Plotting
The next thing we want to do is plot our data.  Here's how to make a simple plot.

```python
# Using the example of y(x)=ax^2+bx+c from above,
# The simplest plot is just plot(x,y)
plot(my_x,my_y)
# TIP: If you want multiple lines on a plot, just call plot() once for every line
```

```python
# Let's make our plot square. We have to define the figsize of our figure.
fig = figure(figsize=(6,6))

# [STYLE]: Now, let's add some complexity to our plot. 
# (1) Colors: We can specify many colors by the first letter of their name:
# b=blue, g=green, r=red, y=yellow, w=white, c=cyan, m=magenta, k=black
# (2) Linestyles: We can also specify line styles:
# -- = dash, : = dotted, -. = dash-dot, - = solid
# (3) Marker Shapes: And we can specify marker styles:
# s=square, *=star, ^=triangle, o=circle
# (Bonus) Combinations:
# We can combo these line/marker parameters together as the third parameter.

# Let's plot the polynomial again, but this time with some style.
plot(my_x,my_y,'r:*')

# Let's put a random marker at the point (2,20)
plot([2],[20],'gs')

# [AXIS LABELS]: Let's also add labels and a title because those are important
# And required on the plots you turn in!
xlabel('x')
ylabel('y')
title('Polynomial plot')

# [AXIS LIMITS] Finally, let's to shift the axes so that we can better see all the points. 
xlim(-0.1,4.1) # xmin,xmax (I'm setting them to be just outside my x range so the end points show up.)
ylim(-5.1,35.1)

# [SAVE FIGURE] Now that we have a pretty plot, we can save our figure using savefig()
savefig('prettyplot.png')
```

# Today's assignment

Our task for today is to plot elliptical orbits.  From class, remember that 
$$ r(\phi)=\frac{a(1-e^2)}{1+e\cos\phi}$$

**(c) The first thing we need to do is create an array of values for $\phi$ that we can then use to calculate $r$.**

```python
phi = arange(0,2*pi+0.01, pi/12)
phi = array(phi)
print(phi)
```

**(d) Try constructing a lambda function yourself for $r(\phi)$ by passing in $a$, $e$, and $\phi$.**<br>
*Note*: in Python, $e$ is a defined constant.  Thus, define the eccentricity variable as something like `ecc` or another name of your choice.

```python
r = lambda a, ecc, ph: a*(1-ecc**2) / (1+ecc*cos(phi)) 
#print(r)
```

**(e) Now, compute $r(\phi)$ for a circular orbit with semi-major axis 1 AU.**

```python
print(r(1,0,phi))
fun1 = r(1,0,phi)
```

**(f) Make a plot of $r(\phi)$ in the $(x,y)$ plane. How do we decompse $r$ into $x$ and $y$? Also, plot the central star on your plot.**

```python
# r = sqrt(x**2+y**2)
# x = r*cos(phi), y = r*sin(phi)
el_x = fun1*cos(phi)
el_y = fun1*sin(phi)
plot(el_x, el_y)
plot([0],[0],'gs')
```

Your plot is pretty boring, right?  Just a circle.

**(g) Now try plotting the orbit of an eccentric planet.**

```python
fun2 = r(1,0.5,phi)
el_x = fun2*cos(phi)
el_y = fun2*sin(phi)
plot(el_x, el_y)
plot([0],[0],'gs')
```

With this power, we can now visualize the 2d orbits for any planet we want.
