```python
from numpy import *
from matplotlib.pyplot import *
%matplotlib inline

g = (4.30091)*10**-3 #pc * M⊙**(−1) * (km/s)**(2)
v = (4.*np.pi*0.1*g)**(1./2.)
t = np.arange(-50, 100)
a = 7. / v
b = 25
z = a*np.sin(v*t) + b*np.cos(v*t)

plot(t, z)
xlabel('Time [millions of years]')
ylabel('Displacement from Galactic Plane [pc]')
title("The Sun's Vertical Motion")


```

```python
from scipy import optimize
from numpy import *

def f(t):
    g = (4.30091)*10**-3 #pc * M⊙**(−1) * (km/s)**(2)
    v = (4.*np.pi*0.1*g)**(1./2.)
    a = 7. / v
    b = 25
    z = a*np.sin(v*t) + b*np.cos(v*t)
    return z

def f_prime(t):
    g = (4.30091)*10**-3 #pc * M⊙**(−1) * (km/s)**(2)
    v = (4.*np.pi*0.1*g)**(1./2.)
    a = 7. / v
    b = 25
    dz = a*v*cos(v*t) - b*v*sin(v*t)
    return dz

optimize.root_scalar(f, x0=-5., x1=-20.)
```
