```python
from numpy import *
from matplotlib.pyplot import *
%matplotlib inline

t = np.arange(0, 0.02, 0.001)

def calcM(r, t):
    Eisco = sqrt(1-2/(3*r))
    m0 = 6*10**9
    eps = 1 - Eisco
    tedd = 0.45 # Gyr
    m = m0*np.e**((1-eps)*t/(eps*tedd))
    print(r, Eisco)
    return m

m1 = calcM(1, t)
m2 = calcM(6, t)

plot(t, m1)
plot(t, m2)
xlabel('Time in Gyr')
ylabel('Mass in Msun')
title("The Mass of blackhole M87* over time")
```
