# Measuring the Read Noise

## CCD Operation:

Notes: We obtained the darks by holding the lens cap over the CCD detector to remove any light from hitting it. We also used CCD Ops to tell the CCD to take Darks only during this part.

### Bias

```python
ls /mnt/c/Users/panda/Documents/ObsAstronomy/ObsData/LabOneData/Bias/
```

```python
b_lis = glob(bias_folder)
bias_array = np.array([fits.getdata(f) for f in b_lis])
master_bias = np.median(np.array(bias_array), axis=0)
```

```python
norm = simple_norm(master_bias, 'log')
plt.imshow(master_bias, norm=norm)
plt.colorbar()
```

### Dark

Notes: We did not take a bias between each dark image.

```python
ls /mnt/c/Users/panda/Documents/ObsAstronomy/ObsData/LabOneData/Dark/
```

```python
d_lis = glob(dark_folder)
dark_array = np.array([fits.getdata(f) for f in d_lis])
darks_mean = []
```

```python
# 0.05 s exposures
lis = glob('/mnt/c/Users/panda/Documents/ObsAstronomy/ObsData/LabOneData/Dark/dark_0p05_*')
lis_array = np.array([fits.getdata(f) for f in lis])
im = np.median(np.array(lis_array), axis=0)-master_bias

norm = simple_norm(im, 'log')
plt.imshow(im, norm=norm)
plt.colorbar()
darks_mean.append(im)
```
